# The Else and Elif Clauses


```python
harga_jeruk=50000
standar_jeruk=30000
#apabila harga jeruk lebih tinggi dari standarnya maka jeruk itu mahal, 
#apabila harga jeruk lebih rendah jeruk itu murah,
#kalau sama jeruk itu biasa

if harga_jeruk>standar_jeruk:
    print('mahal')
elif harga_jeruk<standar_jeruk:
    print('murah')
elif harga_jeruk==standar_jeruk:
    print('biasa')
```

```python
harga_jeruk=int(input('harga jeruk :'))
standar_jeruk=30000
#apabila harga jeruk lebih tinggi dari standarnya maka jeruk itu mahal, 
#apabila harga jeruk lebih rendah jeruk itu murah,
#kalau sama jeruk itu biasa

if harga_jeruk>standar_jeruk:
    print('mahal')
elif harga_jeruk<standar_jeruk:
    print('murah')
else:
    print('biasa')
```

# QUIZ 1
harga pisang 3000, jika belanja lebih dari 100.000 diskon 10%, jika 50.000 sampai 100.000 diskon 5%, dan jika dibawah 50.000 tidak mendapat diskon

```python
jumlah_pisang=int(input('jumlah pisang yang dibeli :'))
harga_pisang=3000
belanja=harga_pisang*jumlah_pisang
if belanja>100000:
    total=belanja-(belanja*0.1)
    print('total belanja anda adalah IDR ' + str(total) + ' (' + str(belanja) + 'di diskon 10%)')
if belanja>50000 and belanja<100000:
    total=belanja-(belanja*0.05)
    print('total belanja anda adalah IDR ' +  str(total) +' (' + str(belanja) + 'di diskon 5%)')
else:
    total=belanja
    print('total belanja anda adalah IDR ' + str(total))
```

# QUIZ 2
buat suatu kondisi dimana jika year service lebih dari 10 tahun, maka akan mendapat bonus 10% dari gaji, jika kurang dari itu tidak mendapat bonus

```python
year_service=int(input('year of service'))
gaji=int(input('berapa gaji anda'))
if year_service>10:
    total_gaji=gaji+(gaji*0.10)
    print('Selamat anda mendapat bonus! Total gaji anda IDR'+str(total_gaji))
else:
    print('Maaf anda tidak mendapatkan bonus! Total gaji anda IDR'+str(gaji))
```

# While Loop

```python
n=5
while n>0:
    n-=1
    print(n)
```

```python
apel=5
apel-=1
print(apel)
```

```python
apel=5
while apel>0:
    apel-=1
    if apel==1:
        print('woy tinggal 1 nih')
    else:
        print(apel)
```

```python
# break digunakan untuk memberhentikan loop
apel=5
while apel>0:
    apel-=1
    if apel==2:
        break
    print(apel)
print('loop ended')
```

```python
apel
```

```python
# continue digunakan melanjutkan loop
apel=5
while apel>0:
    apel-=1
    if apel==2:
        continue
    print(apel)
print('loop ended')
#(ngeprint loop ended ketika proses loopingnya selesai)
```

```python
# else
apel=10
while apel>0:
    apel-=2
    print(apel)
else:
    print('loop ended')
#(ngeprint loop ended ketika di while sudah false)
```

```python
apel=5
while apel>0:
    apel-=1
    print(apel)
    if apel==2:
        break
else:
    print('loop ended')
#kondisi while tidak pernah dilanggar, saat apel=2 keburu berhenti, jadi gaada loop ended
```

# Nested If - Else & Loops

```python
#Nested if
umur=40
gender='p'
if umur<18:
    if gender=='p':
        print('son')
    else:
        print('daughter')
if umur>=18 and umur<65:
    if gender=='m':
        print('dad')
    else:
        print('mom')
```

```python
# continue : syntax untuk for
for i in['Aster','Jubah','Nambo','hachi']:
    if 'b' in i:
        continue
    print(i)
    
#dilooping ada continue : melanjutkan ke index selanjutnya
#ketika ganemu 'b' dilanjutkan ke index berikutnya
```

```python
# break
for i in['Aster','Jubah','Nambo','hachi']:
    if 'b' in i:
        break
    print(i)
    
#ketika ganemu 'b' keseluruhan loop diberhentikan
```

```python
# pass : operasi kosong, syntax untuk if
for i in['Aster','Jubah','Nambo','hachi']:
    if 'b' in i:
        pass
    print(i)
```

##### loop itu keseluruhan iterasi objek
* 1 loop diatas ada 4 iterasi (per aster, jubah, nmabo, hachi)
* iterasi adalah istilah untuk 1 putaran loop
* continue : akan memberhentikan iterasi dan dilanjutkan ke iterasi berikutnya
* break : akan memberhentikan looping secara keseluruhan

# QUIZ 3
buatlah while loop, ketika ganjil maka print angkanya, ketika genap maka print 'genap', batas angka 1-20

```python
i=1
while i<=20:
    if i%2==1:
        print(i)
    else:
        print('genap')
    i+=1
```

```python
tinggi_segitiga = int(input('Input tinggi segitiga: '))
print()
  
for i in range(1,tinggi_segitiga+1):
  for j in range(i):
    print(i,' ',end='',sep='')
  print()
```
